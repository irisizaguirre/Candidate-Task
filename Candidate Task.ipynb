```python
import numpy as np
import pandas as pd

#load dataset and information
dataset_insurance = pd.read_csv("dataset.csv")
dataset_insurance.info()
```

```python
#Visualization of the table head
dataset_insurance.head()
```

```python
#Sub-dataset: Filtering "Billing City"object with null entries
dataset_insurance_bc = dataset_insurance[dataset_insurance["Billing City"].isnull()]
dataset_insurance_bc.info()
```

```python
#Sub-dataset: Filtering "Billing City"object with null entries
# Number of account statuses and sum of accounts by account status
dataset_insurance_as = dataset_insurance_bc.sort_values('Account ID').groupby('Account Status', as_index=False).agg(
    {'Account Name': pd.Series.nunique})
dataset_insurance_as
```

```python
#Sum of account ID by Account Owner
owner_table = dataset_insurance.sort_values('Account ID').groupby('Account Owner', as_index=False).agg(
    {'Account ID': pd.Series.nunique})
owner_table
```
